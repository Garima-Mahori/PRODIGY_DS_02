### OCCUPATION 

```python
# Group by the 'occupation' column and calculate the mean of the 'weight' column
average_weight_by_occupation = df.groupby('OccupationType')['Weight'].mean().reset_index()

# Display the result
print(average_weight_by_occupation)

# Plot the average weights
plt.figure(figsize=(6, 4))
plt.bar(average_weight_by_occupation['OccupationType'], average_weight_by_occupation['Weight'], color='skyblue')
plt.xlabel('Occupation')
plt.ylabel('Average Weight')
plt.title('Average Weight by Occupation')
plt.xticks(rotation=45)
plt.show()
```

# Multivariate Analysis

```python
numeric_df = df.select_dtypes(include=[float, int])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
# Scatter matrix using pandas
pd.plotting.scatter_matrix(df, figsize=(12, 8))
plt.show()
```

```python
# Pivot the data
pivot_table = df.pivot_table(index='AgeGrp', columns='Gender', aggfunc='size', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title('AGE vs GENDER Data Heatmap')
plt.show()
```
